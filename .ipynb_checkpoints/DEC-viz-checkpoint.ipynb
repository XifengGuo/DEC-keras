```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from DEC import DEC
from datasets import load_mnist
```

```python
import glob
import numpy as np
from PIL import Image
filelist = glob.glob('/path/to/images')
x = np.array([np.array(Image.open(fname).resize((100,100), Image.ANTIALIAS), dtype=object) for fname in filelist], dtype=object)
x = x.reshape((x.shape[0], -1))
x = np.divide(x, 255.).astype('float32')
```

```python
x.shape
```

```python
dec = DEC(dims=[x.shape[-1], 500, 500, 2000, 5])
dec.load_weights('results/DEC_model_final.h5')
```

```python
dec.encoder.summary()
```

```python
p = dec.predict(x)
```

```python
count_arr = np.bincount(p)
for i in range(5):
    print(i, count_arr[i])
```

```python
G = len(np.unique(p)) #Number of labels#2D matrix  for an array of indexes of the given label
cluster_index= [[] for i in range(G)]
for i, label in enumerate(p,0):
    for n in range(G):
        if label == n:
            cluster_index[n].append(i)
        else:
            continue#Visualisation for clusters = clustplt.figure(figsize=(20,20));
plt.figure(figsize=(20,20));
clust = 0 #enter label number to visualise
num = 48 #num of data to visualize from the cluster
for i in range(1,num): 
    plt.subplot(10, 10, i); #(Number of rows, Number of column per row, item number)
    plt.imshow(x[cluster_index[clust][i]].reshape(100, 100 ,3));
    
plt.show()
```
